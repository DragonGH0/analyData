```python
import pandas as pd
import numpy as np
from pandas import read_csv
```

```python
filename="basic.csv"
f=open(filename,encoding='UTF-8')
data=read_csv(f)
df = pd.DataFrame(data)
```

```python
df = df.drop(df.columns[0],axis=1)
```

```python
# del columns
df = df.drop(columns=['cityName','city_confirmedCount','city_curedCount','city_deadCount'])
```

```python
df.tail(20)
```

```python
# 删除无用列
count = 0
s_index = pd.Series()  # 存储有用列的index
# 用每天都在改变的治愈数筛选有用index
for i in range(len(df)):
    if df.iloc[i,2] != count:
        s_index = s_index.append(pd.Series(i),ignore_index=True)
        count = df.iloc[i,2]
```

```python
df = df[df.index.isin(s_index)]
df
```

```python
df.shape
```

```python
df.index = np.arange(65)
```

```python
# after  cut row
s_date = pd.date_range(start='2020-01-25',end='2020-03-29')
s_date.name = 'updateTime'
df.iloc[:,4] = s_date
```
