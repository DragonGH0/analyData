```python
import pandas as pd
import numpy as np
```

```python
from pandas import read_csv
filename="DXYArea.csv"
f=open(filename,encoding='UTF-8')
data=read_csv(f)
```

```python
df = pd.DataFrame(data)
```

```python
df.columns
```

```python
# 删除无用列
df = df[df["provinceName"].isin(["湖北省"])]
df = df.drop(["continentName","continentEnglishName","countryName","countryEnglishName"],axis=1)
df = df.drop(columns=['provinceEnglishName','province_zipCode','province_suspectedCount','cityEnglishName','city_suspectedCount'])
df = df.drop(columns=['city_zipCode'])
```

```python
df.head(1)
```

```python
# 重新定义索引
df.shape
df.index = np.arange(6549)
```

```python
# 定义有效信息表 df1
df1 = pd.DataFrame(index=np.arange(1),columns=df.columns)
```

```python
# 删除重复日期

# 1简化查询： 简化信息表
df_brief = pd.DataFrame(columns=['updateTime'],data=df['updateTime'])
```

```python
# 2获取每天最后时刻的索引 保存在 dateIndex中
dateIndex = pd.Series()
# 三月
i_date = 30
while i_date >= 2:
    if i_date >= 10:
        s_date = '2020-03-'
    else:
        s_date = '2020-03-0'
    date = s_date + str(i_date)
    # 获得当日所有市的索引（前17个）  并且转化为列表类型
    s = pd.Series(df_brief[df_brief['updateTime'] < date].head(17).index)
    dateIndex = dateIndex.append(s,ignore_index=True)  # 添加到dateIndex
    i_date -= 1
# 二月
i_date = 30
while i_date >= 2:
    if i_date >= 10:
        s_date = '2020-02-'
    else:
        s_date = '2020-02-0'
    date = s_date + str(i_date)
    # 获得当日所有市的索引（前17个）  并且转化为列表类型
    s = pd.Series(df_brief[df_brief['updateTime'] < date].head(17).index)
    dateIndex = dateIndex.append(s,ignore_index=True)  # 添加到dateIndex
    i_date -= 1
i_date = 32
# 一月
while i_date >= 26:
    s_date = '2020-01-'
    date = s_date + str(i_date)
    # 获得当日所有市的索引（前17个）  并且转化为列表类型
    s = pd.Series(df_brief[df_brief['updateTime'] < date].head(17).index)
    dateIndex = dateIndex.append(s,ignore_index=True)  # 添加到dateIndex
    i_date -= 1

# print(dateIndex.head(17))
# print(dateIndex.tail(17))
```

```python
df[df.index.isin(dateIndex)].head(17)
```

```python
# 3删除重复的日期信息
df = df[df.index.isin(dateIndex)]
df.shape
```

```python
# 重新排序 定义索引
df = df.iloc[::-1]
df.index = np.arange(1105)
```
